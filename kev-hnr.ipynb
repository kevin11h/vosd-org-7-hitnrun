# VoSD Hit and Run Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
from  sdipylib.url import cache_url
%matplotlib inline
```

```python
fn = cache_url('http://devtest.sandiegodata.org.s3.amazonaws.com/manifests/7a007ac7-960f-468b-8ce3-7b9f83e34ea8/hnr0914.csv')
hnr_0914 = pd.read_csv(fn, parse_dates=['offense_date'])
hnr_0912 = hnr_0914[(2009 <= hnr_0914.complaint_year) & (hnr_0914.complaint_year <= 2012)]

print "There are {0} number of cases in total from years 2009 to 2012, inclusive,".format(len(hnr_0912))
print "out of {0} number of cases in total from years 2009 to 2014, inclusive.".format(len(hnr_0914))
```

##  2009-2012 Closed Cases where PC187, PC191, PC192, or VC20001 is the Highest Charge

<i><b>Note:</b>"About final disposition: Please exclude dismissed cases, and blanks ... these cases are still in development."</i>

#### From a quick glance at the chart below, VC20001 is the most common among the other PC187, 191, and 192 charges - nearly ten times as the rest of all PCs combined.

```python
is_PC187 = (hnr_0912.highest_charge.str.contains("PC187"))
is_PC191 = (hnr_0912.highest_charge.str.contains("PC191"))
is_PC192 = (hnr_0912.highest_charge.str.contains("PC192"))
is_VC20001 = (hnr_0912.highest_charge.str.contains("VC20001"))
matched_cases = hnr_0912[is_PC187 ^ is_PC191 ^ is_PC192 ^ is_VC20001]

is_closed = ~pd.isnull(hnr_0912.final_disposition) # ensure that the final disposition field is not missing
matched_cases_that_are_closed = hnr_0912[is_closed & (is_PC187 ^ is_PC191 ^ is_PC192 ^ is_VC20001)]
```

```python
print "There are {0} total PC187, PC191, PC192, and VC20001 cases from years 2009 to 2012, inclusive.".format(len(matched_cases))
print "There are {0} closed PC187, PC191, PC192, and VC20001 cases from years 2009 to 2012, inclusive.".format(len(matched_cases_that_are_closed))

penal_codes = ("PC187", "PC191", "PC192", "VC20001")
penal_code_counts = map(lambda condition : sum(condition & is_closed),[is_PC187, is_PC191, is_PC192, is_VC20001])
penal_code_counts_sorted = sorted(penal_code_counts, reverse=True)
penal_codes_by_counts = dict(zip(penal_code_counts, penal_codes))
penal_codes_sorted = [penal_codes_by_counts[cnt] for cnt in penal_code_counts_sorted]

# bar chart
x_pos = range(len(penal_codes))
plt.bar(x_pos, penal_code_counts_sorted, align='center', alpha=0.4)
plt.xticks(x_pos, penal_codes_sorted)
plt.ylabel("Number of Cases",labelpad=20)
plt.title("Closed Cases from 2009 to 2012\nHighest Charge Breakdown")

# simple style
plt.gcf().set_size_inches(12,9)
plt.gcf().set_facecolor('white')
plt.gca().yaxis.grid(color="white", linestyle="-")
plt.tick_params(axis='x',bottom='off',top='off')
plt.tick_params(axis='y',left='off',right='off')

# annotate
for x, c in zip(x_pos, penal_code_counts_sorted):
    plt.annotate("{0}".format(c), (x, c+2), size='small', ha='center')
```

## Incarceration Frequency, Yearly Summary

#### It may be coincidental, but according to the multiple bar chart, the yes incarceration vs no incarceration ratio is almost evenly split, but 2012 may be a bit in favor of incarceration than years 2009 to 2011 however.

```python
cases = matched_cases_that_are_closed.fillna(0)
cases["has_incar_time"] = 0+((cases.incar > 0) | (cases.incar_county > 0) | (cases.incar_life > 0) | (cases.incar_prison > 0))
cases["nil_incar_time"] = 1-cases["has_incar_time"]

highest_charge_to_raw_penal_code_mapping = (lambda hc : (filter(lambda pc : pc in hc, penal_codes)[0]))
cases["raw_penal_code"] = cases.highest_charge.apply(highest_charge_to_raw_penal_code_mapping)
```

```python
pd.options.display.mpl_style = 'default'

incarceration_data = cases[["raw_penal_code", "has_incar_time", "nil_incar_time"]].copy()
                
for y in range(2008, 2012):
    year = y+1
    incarceration_info_by_year = incarceration_data[cases.complaint_year == year]
    groupby_raw_penal_code = incarceration_info_by_year.groupby("raw_penal_code")
    
    incarceration_frequency_by_penal_code = groupby_raw_penal_code.sum()        
    incarceration_frequency_by_penal_code.index.names = [""]
    incarceration_frequency_by_penal_code.columns = ["Incarceration", "No Incarceration"]
    for pc in penal_codes:
        if pc not in incarceration_frequency_by_penal_code.index:
            incarceration_frequency_by_penal_code.loc[pc] = [0, 0]
    incarceration_frequency_by_penal_code.ix[penal_codes_sorted].plot(kind="bar",
                                                                      rot=0,
                                                                      xlim=(0,len(penal_codes)),
                                                                      ylim=(0,70))
    # annotate
    total_n_cases = 0
    
    for i, pc in enumerate(penal_codes_sorted):
        y_incar = int(incarceration_frequency_by_penal_code.ix[pc]["Incarceration"])
        n_incar = int(incarceration_frequency_by_penal_code.ix[pc]["No Incarceration"])
        plt.annotate("{0}".format(y_incar), (i+.4,y_incar+1))
        plt.annotate("{0}".format(n_incar), (i+.8,n_incar+1))
        total_n_cases += (y_incar + n_incar)
        
    plt.title("{0} Incarceration Sentencings (Frequency = {1})".format(year, total_n_cases))
    plt.ylabel("Number of Cases")
    plt.gcf().set_size_inches(8,6)
```

## Average Incarceration Length, from 2009-2012

#### According to the five-number summary, there are 130 counts of actual incarceration for charges of PC187, PC191, PC192, and VC20001 from years 2009 to 2012, each lasting on average 854 days (roughly 2 years).<br /> However, in terms of days, these statistics are highly skewed! This goes to show that scaling time based on days or years makes a huge difference in perception. The standard deviation is unreasonably close to 2000 days (which is around 5 years).

```python
cases_with_incarceration_09_to_12 = cases[cases["has_incar_time"] == True]
incarceration_df = cases_with_incarceration_09_to_12[["incar"]].copy()
incarceration_df.columns = ["Incarceration"]
incarceration_df.describe()
```

#### In a time scale of years, the data seems a bit more reasonable, aside from the obvious problem of outliers. Let's count how many outliers there are and separate them from the main cluster.

```python
DAYS_IN_YEAR = 365 # omit leap years for now
inyears = incarceration_df.describe()[1:]/DAYS_IN_YEAR
inyears
```

```python
q3 = incarceration_df.describe().ix["75%"].values[0]
q1 = incarceration_df.describe().ix["25%"].values[0]
iqr = q3 - q1
is_outlier = (incarceration_df.Incarceration.values < (q1 - 1.5*iqr)) | (incarceration_df.Incarceration.values > (q3 + 1.5*iqr))
main_cluster_incarceration_df = incarceration_df[~is_outlier]

main_cluster_incarceration_df_inyears = main_cluster_incarceration_df/365
main_cluster_incarceration_df_inyears.boxplot(); plt.gcf().set_size_inches(9,12);
main_cluster_incarceration_df_inyears.describe()
```

#### We've dropped around a 20% of the original data, which were outliers, and now we have 110 counts of incarceration for charges of PC187, PC191, PC192, and VC20001 from years 2009 to 2012, each lasting on average 275 days (roughly 3/4 of the year).<br />

```python
outliers_df = incarceration_df[is_outlier]
outliers_df_in_years = (outliers_df/365)
outliers_df_in_years.boxplot()
outliers_df_in_years.describe()
```

#### And now we've discovered the other 20% which are the outliers, and there incarceration time seems more punitative with an average of 10 years! A huge difference from the first population which had only an average of 1 year of incarceration time. There is something else that may not be accounted for this glaring discrepancy.
